```python
import matplotlib.pyplot as plt

regioes = ["Norte", "Nordeste", "Centro-Oeste", "Sudeste", "Sul"]
populacao = [18.4, 54.6, 16.3, 89.6, 29.9]

plt.bar(regioes, populacao, color="steelblue")
plt.title("Populacao do Brasil por Regiao")
plt.xlabel("Regiao")
plt.ylabel("Milhoes de habitantes")
plt.tight_layout()
plt.show()
```
